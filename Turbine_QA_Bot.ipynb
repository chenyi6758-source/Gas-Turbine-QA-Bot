```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("📚 正在加载《燃气轮机维护手册》及初始化 AI 语言引擎...\n")

# 1. 模拟企业的《设备维护与故障排查手册》（英文，因为外企基本用英文文档）
manual_rules = [
    "Check the air filter every 500 hours of operation to prevent compressor stall.", # 规则0：每500小时检查空气过滤器，防止压气机失速。
    "The combustion chamber must be inspected for cracks during the annual maintenance.", # 规则1：年度维护时必须检查燃烧室是否有裂纹。
    "If the exhaust temperature exceeds 600 degrees, shut down the turbine immediately.", # 规则2：如果排气温度超过600度，立即停机。
    "Lubricating oil pressure should be maintained between 2.5 and 3.5 bar.", # 规则3：润滑油压力应保持在2.5到3.5巴之间。
    "Vibration sensors should be calibrated every 6 months to ensure accurate rotor balancing." # 规则4：振动传感器应每6个月校准一次。
]

# 2. 召唤 NLP 算法：把人类语言转换成电脑能懂的“数学向量” (TF-IDF 算法)
vectorizer = TfidfVectorizer()
# 让 AI 通读并“背诵”这本手册
knowledge_base_vectors = vectorizer.fit_transform(manual_rules)

print("✅ AI 助手已就绪，已熟读全部维护规则！\n")
print("-" * 50)

# 3. 核心功能：定义一个“问答机器人”函数
def ask_turbine_bot(user_question):
    print(f"👨‍🔧 工程师提问: {user_question}")
    
    # 将用户的问题也转换成数学向量
    question_vector = vectorizer.transform([user_question])
    
    # 计算问题和手册里每一句话的“相似度”
    similarities = cosine_similarity(question_vector, knowledge_base_vectors)
    
    # 找到最匹配的那一条规则的编号
    best_match_index = similarities.argmax()
    highest_score = similarities[0][best_match_index]
    
    # 如果得分太低，说明手册里没写
    if highest_score < 0.1:
        print("🤖 AI 回答: 对不起，维护手册中没有找到与此相关的内容，请联系高级工程师。")
    else:
        print(f"🤖 AI 回答: 根据手册规定 -> {manual_rules[best_match_index]}")
        print(f"   (AI 匹配信心指数: {highest_score * 100:.1f}%)")
    print("-" * 50)

# 4. 实战演练：工程师开始提问！
# 提问一：关于温度过高的紧急情况
ask_turbine_bot("What should I do if the temperature is over 600?")

# 提问二：关于保养频率
ask_turbine_bot("How often do we need to check the filter?")

# 提问三：瞎问一个无关的问题
ask_turbine_bot("Where is the nearest cafeteria?")
```
